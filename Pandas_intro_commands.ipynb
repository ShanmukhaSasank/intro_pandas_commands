```python
# prompt: can you import basic library pandas in this workbook

import pandas as pd

# Assuming Salaries.csv is in the current working directory.
# If not, provide the correct path to the file.
try:
    df = pd.read_csv('Salaries.csv')
    print(df)
    print(df.head())
    print(df.tail())
    print(df.columns)
    print(df.shape)
    print(df.dtypes)
    print(df.info())
    pd.set_option('display.max_rows', None)
    print(df)
    print(df.isnull().sum())
    print(df.describe(include='all'))
    print(df['rank'].value_counts())
    print(df.mean(numeric_only=True))
    print(df['phd'].mean())
    print(df['phd'].median())
    print(df['salary'].mean())
    print(df['salary'].median())
    print(df['salary'].mode())
    print(df[['phd', 'salary']].mean())
    print(df['phd'].var())
    print(df[['phd', 'salary']].std())
    print(df[['phd', 'salary']].skew())
    print(df[['phd', 'salary']].kurt())
    print(df[['phd', 'salary']].max() - df[['phd', 'salary']].min())
    print(df[['phd', 'salary']].describe())
    print(df['salary'].count())
    print(df['rank'].value_counts())
except FileNotFoundError:
    print("Error: Salaries.csv not found. Please upload the file or provide the correct path.")
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
